# Taller Metodos Numericos 

## a) e^x-1

```python
import math

h=input("Ingrese el Valor de h")
xi=-5
print "   xi     Hacia Adelante       Hacia Atras            Centrada       "
print"--------------------------------------------------------------------"
while xi<=5:
    xi_=xi-h
    x1=xi+h
    def fx(x):
        return math.exp(x) -1
    
    def dfx(x):
        return math.exp(x)
        if math.exp(x)==0:
            return 0.00000000001


    def fi(x,x1):
        f1=(fx(x1)-fx(xi))/h
        return f1

    def fm(x,xl):
        f2=(fx(xi)-fx(xi_))/h
        return f2

    def fc(x,xl):
        f3=(fx(x1)-fx(xi_))/(2*h)
        return f3

    print " ",xi," | ",fi(xi,x1)," | ",fm(xi,x1)," | ",fc(xi,x1)," | "
    print "Absoluto | ",math.fabs(dfx(xi)-fi(xi,x1))," | ",math.fabs(dfx(xi)-fm(xi,x1))," | ",math.fabs(dfx(xi)-fc(xi,x1))," | "
    print "Relativo | ", math.fabs((dfx(xi)-fi(xi,x1))/dfx(xi))," | ", math.fabs((dfx(xi)-fm(xi,x1))/dfx(xi))," | ", math.fabs((dfx(xi)-fc(xi,x1))/dfx(xi))," | "
    print "Porcentual | ",math.fabs((dfx(xi)-fi(xi,x1))/dfx(xi))*100," | ",math.fabs((dfx(xi)-fm(xi,x1))/dfx(xi))*100," | ",math.fabs((dfx(xi)-fc(xi,x1))/dfx(xi))*100," | "
    
    #print "hacia adelante: ",fi(xi,x1)
    #print "hacia atras: ",fm(xi,x1)
    #print "centrada: ",fc(xi,x1)
    print "------------------------------------------------------------------------"
    xi+=h
```

# b) Sen(2x)^3+x^2

```python
import math

h=input("Ingrese el Valor de h")
xi=-5
print "   xi     Hacia Adelante       Hacia Atras            Centrada       "
print"--------------------------------------------------------------------"
while xi<=5:
    xi_=xi-h
    x1=xi+h
    def fx(x):
        return ((math.sin(2*x))**3)+(x**2)
    
    def dfx(x):
        auxil=(2*(x+(3*((math.sin(2*x))**2)*(math.cos(2*x)))))
        if auxil==0:
            return (0.0000000001)
        return auxil

    def fi(x,x1):
        f1=(fx(x1)-fx(xi))/h
        return f1

    def fm(x,xl):
        f2=(fx(xi)-fx(xi_))/h
        return f2

    def fc(x,xl):
        f3=(fx(x1)-fx(xi_))/(2*h)
        return f3

    print " ",xi," | ",fi(xi,x1)," | ",fm(xi,x1)," | ",fc(xi,x1)," | "
    print "Absoluto | ",math.fabs(dfx(xi)-fi(xi,x1))," | ",math.fabs(dfx(xi)-fm(xi,x1))," | ",math.fabs(dfx(xi)-fc(xi,x1))," | "
    print "Relativo | ", math.fabs((dfx(xi)-fi(xi,x1))/dfx(xi))," | ", math.fabs((dfx(xi)-fm(xi,x1))/dfx(xi))," | ", math.fabs((dfx(xi)-fc(xi,x1))/dfx(xi))," | "
    print "Porcentual | ",math.fabs((dfx(xi)-fi(xi,x1))/dfx(xi))*100," | ",math.fabs((dfx(xi)-fm(xi,x1))/dfx(xi))*100," | ",math.fabs((dfx(xi)-fc(xi,x1))/dfx(xi))*100," | "
    
    #print "hacia adelante: ",fi(xi,x1)
    #print "hacia atras: ",fm(xi,x1)
    #print "centrada: ",fc(xi,x1)
    print "------------------------------------------------------------------------"
    xi+=h

```

# c) x^4+3x^2-cos(x)

```python
import math

h=input("Ingrese el Valor de h")
xi=-5
print "   xi     Hacia Adelante       Hacia Atras            Centrada       "
print"--------------------------------------------------------------------"
while xi<=5:
    xi_=xi-h
    x1=xi+h
    def fx(x):
        return ((x**4)+(3*x**2)-(math.cos(x)))
    
    def dfx(x):
        auxil=((4*x**3)+(6*x)+(math.sin(x)))
        if auxil==0:
            return (0.0000000001)
        return auxil
    def fi(x,x1):
        f1=(fx(x1)-fx(xi))/h
        return f1

    def fm(x,xl):
        f2=(fx(xi)-fx(xi_))/h
        return f2

    def fc(x,xl):
        f3=(fx(x1)-fx(xi_))/(2*h)
        return f3

    print " ",xi," | ",fi(xi,x1)," | ",fm(xi,x1)," | ",fc(xi,x1)," | "
    print "Absoluto | ",math.fabs(dfx(xi)-fi(xi,x1))," | ",math.fabs(dfx(xi)-fm(xi,x1))," | ",math.fabs(dfx(xi)-fc(xi,x1))," | "
    print "Relativo | ", math.fabs((dfx(xi)-fi(xi,x1))/dfx(xi))," | ", math.fabs((dfx(xi)-fm(xi,x1))/dfx(xi))," | ", math.fabs((dfx(xi)-fc(xi,x1))/dfx(xi))," | "
    print "Porcentual | ",math.fabs((dfx(xi)-fi(xi,x1))/dfx(xi))*100," | ",math.fabs((dfx(xi)-fm(xi,x1))/dfx(xi))*100," | ",math.fabs((dfx(xi)-fc(xi,x1))/dfx(xi))*100," | "
    
    #print "hacia adelante: ",fi(xi,x1)
    #print "hacia atras: ",fm(xi,x1)
    #print "centrada: ",fc(xi,x1)
    print "------------------------------------------------------------------------"
    xi+=h

```
