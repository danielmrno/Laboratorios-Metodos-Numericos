```python
import math
import numpy as np
import pylab as pl
ycalculada=[10,16.3,23,27.5,31,35.6,39,41.5,42.9,45,46,45.5,46,47,50]
valoresx1 = []
valoresy1 = []
x=[]
y=[]
sumatoriaxiyi=0.0
sumatoriaxi=0.0
sumatoriayi=0.0
sumatoriaxicuad=0.0
aux=1
m=68.1
g=9.8
c=12.5

def fx(a):
    return ((g*m)/c)*(1-math.exp((-c/m)*a))
    
while (aux<=15):
    y0=fx(aux)
    y.append(y0)
    x.append(aux)
    aux+=1

yi=np.array(y)
xi=np.array(x)
aux=0
while (aux<15):
    sumatoriaxiyi+=(yi[aux]*xi[aux])
    sumatoriaxi+=xi[aux]
    sumatoriayi+=yi[aux]
    sumatoriaxicuad+=xi[aux]**2
    aux+=1
ai=0.0
ai= ((20*sumatoriaxiyi)-(sumatoriaxi*sumatoriayi))/((20*sumatoriaxicuad)-(sumatoriaxi**2))
print "El valor de ai es: ",ai
a0=0.0
a0=yi.mean()-(ai*xi.mean())
print "El valor de a0 es: ",a0

x=1.0
def fx1(x):
    return a0+(ai*x)

while (x<15):
    y1=fx1(x)
    valoresx1.append(x)
    valoresy1.append(y1)
    x+=1
x=0.0

f1 = pl.plot(yi, '.')
f2=  pl.plot(valoresx1, valoresy1, "r")
f3= pl.plot(xi,ycalculada,'x')

pl.title("Modelo 1")
pl.xlabel("x")
pl.ylabel("y")
f=["Modelo","Regresion","Medido"]
pl.legend(f,loc=4)
pl.xlim(0, 15)
pl.ylim(0, 50)
pl.grid(True)


pl.show()
```

```python
import math
import numpy as np
import pylab as pl
ycalculada=[10,16.3,23,27.5,31,35.6,39,41.5,42.9,45,46,45.5,46,47,50]
valoresx1 = []
valoresy1 = []
x=[]
y=[]
sumatoriaxiyi=0.0
sumatoriaxi=0.0
sumatoriayi=0.0
sumatoriaxicuad=0.0
aux=1
m=68.1
g=9.8
c=12.5

def fx(a):
    return ((g*m)/c)*(a/(3.75+a))
    
while (aux<=15):
    y0=fx(aux)
    y.append(y0)
    x.append(aux)
    aux+=1

yi=np.array(y)
xi=np.array(x)
aux=0
while (aux<15):
    sumatoriaxiyi+=(yi[aux]*xi[aux])
    sumatoriaxi+=xi[aux]
    sumatoriayi+=yi[aux]
    sumatoriaxicuad+=xi[aux]**2
    aux+=1
ai=0.0
ai= ((20*sumatoriaxiyi)-(sumatoriaxi*sumatoriayi))/((20*sumatoriaxicuad)-(sumatoriaxi**2))
print "El valor de ai es: ",ai
a0=0.0
a0=yi.mean()-(ai*xi.mean())
print "El valor de a0 es: ",a0

x=1.0
def fx1(x):
    return a0+(ai*x)

while (x<15):
    y1=fx1(x)
    valoresx1.append(x)
    valoresy1.append(y1)
    x+=1
x=0.0

f1 = pl.plot(yi,'.')
f2=  pl.plot(valoresx1, valoresy1, "r")
f3= pl.plot(xi,ycalculada,'x')

pl.title("Modelo 2")
pl.xlabel("x")
pl.ylabel("y")
pl.xlim(0, 15)
pl.ylim(0, 50)
pl.grid(True)
f=["Modelo","Regresion","Medido"]
pl.legend(f,loc=4)

pl.show()
```
