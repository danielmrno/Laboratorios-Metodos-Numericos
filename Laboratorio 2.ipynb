```python
import math
n=4
error=100
estimado=(0.5*(10**(2-n)))
m=68.1
c=12.5
g=9.8
va=0
t=0
calc=0
print "Tiempo       Valor       Valor Calculado       error"
while error > estimado:
    f=1-math.exp(-(c/m)*(t))
    va=calc
    calc = va + (g-((c*va)/m))
    v=((m*g)/c)*f
    error=(calc-v)
    print "  ",t,"    ",v,"    ",calc,"    ", error
    t +=1
```
